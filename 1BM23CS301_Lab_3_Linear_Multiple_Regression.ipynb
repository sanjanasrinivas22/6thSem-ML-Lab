```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("canada_per_capita_income.csv")

X = df[['year']]
y = df['per capita income (US$)']

model = LinearRegression()
model.fit(X, y)

test = pd.DataFrame([[2020]], columns=['year'])
prediction = model.predict(test)

print(prediction[0])
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("canada_per_capita_income.csv")

x = df['year'].values
y = df['per capita income (US$)'].values

n = len(x)

b1 = (n*np.sum(x*y) - np.sum(x)*np.sum(y)) / (n*np.sum(x*x) - (np.sum(x))**2)
b0 = (np.sum(y) - b1*np.sum(x)) / n

pred = b0 + b1*2020
print("Predicted income:", pred)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("hiring.csv")

mapping = {
    'zero':0,'one':1,'two':2,'three':3,'four':4,
    'five':5,'six':6,'seven':7,'eight':8,'nine':9,
    'ten':10,'eleven':11,'twelve':12
}

df['experience'] = df['experience'].map(mapping)
df['experience'] = df['experience'].fillna(0)
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].mean())

X = df[['experience','test_score(out of 10)','interview_score(out of 10)']]
y = df['salary($)']

model = LinearRegression()
model.fit(X,y)

test1 = pd.DataFrame([[2,9,6]], columns=X.columns)
test2 = pd.DataFrame([[12,10,10]], columns=X.columns)

print(model.predict(test1)[0])
print(model.predict(test2)[0])
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("hiring.csv")

mapping = {
    'zero':0,'one':1,'two':2,'three':3,'four':4,
    'five':5,'six':6,'seven':7,'eight':8,'nine':9,
    'ten':10,'eleven':11,'twelve':12
}

df['experience'] = df['experience'].map(mapping)
df = df.fillna(0)

X = df[['experience','test_score(out of 10)','interview_score(out of 10)']].values
y = df['salary($)'].values

X = np.c_[np.ones(X.shape[0]), X]

beta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

def predict(a,b,c):
    return beta[0] + beta[1]*a + beta[2]*b + beta[3]*c

print(predict(2,9,6))
print(predict(12,10,10))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("1000_Companies.csv")

le = LabelEncoder()
df['State'] = le.fit_transform(df['State'])

X = df[['R&D Spend','Administration','Marketing Spend','State']].values
y = df['Profit'].values

X = np.c_[np.ones(X.shape[0]), X]

beta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

state = le.transform(['Florida'])[0]

pred = beta[0] + beta[1]*91694.48 + beta[2]*515841.3 + beta[3]*11931.24 + beta[4]*state
print("Predicted profit:", pred)
```
