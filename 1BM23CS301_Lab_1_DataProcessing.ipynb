```python
import pandas as pd

# i. To load .csv file into the data frame
df = pd.read_csv("housing.csv")

# ii. To display information of all columns
print(df.info())

# iii. To display statistical information of all numerical columns
print(df.describe())

# iv. To display the count of unique labels for "Ocean Proximity" column
print(df["ocean_proximity"].value_counts())

# v. To display which attributes have missing values count > 0
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
from scipy import stats

diabetes = pd.read_csv("diabetes.csv")

# ------------------ Separate numeric and categorical ------------------
num_cols = diabetes.select_dtypes(include=['int64','float64']).columns
cat_cols = diabetes.select_dtypes(include=['object']).columns

# ------------------ Handle missing values ------------------
diabetes[num_cols] = diabetes[num_cols].replace(0, np.nan)
diabetes[num_cols] = diabetes[num_cols].fillna(diabetes[num_cols].mean())

# ------------------ Encode categorical columns ------------------
le = LabelEncoder()
for col in cat_cols:
    diabetes[col] = le.fit_transform(diabetes[col])

# ------------------ Handle outliers ------------------
z_scores = np.abs(stats.zscore(diabetes[num_cols]))
diabetes = diabetes[(z_scores < 3).all(axis=1)]

# ------------------ Min-Max Scaling ------------------
minmax = MinMaxScaler()
diabetes_minmax = pd.DataFrame(minmax.fit_transform(diabetes),
                                columns=diabetes.columns)

# ------------------ Standard Scaling ------------------
standard = StandardScaler()
diabetes_standard = pd.DataFrame(standard.fit_transform(diabetes),
                                  columns=diabetes.columns)

print("Diabetes preprocessing completed successfully")
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
from scipy import stats

# Load dataset
adult = pd.read_csv("adult.csv")

# ------------------ Data Cleaning ------------------

# Replace '?' with NaN
adult = adult.replace("?", np.nan)

# Fill missing values with mode (safe way, no warning)
for col in adult.columns:
    if adult[col].isnull().sum() > 0:
        adult[col] = adult[col].fillna(adult[col].mode()[0])

# ------------------ Handling Categorical Data ------------------

categorical_cols = adult.select_dtypes(include=['object']).columns
le = LabelEncoder()
for col in categorical_cols:
    adult[col] = le.fit_transform(adult[col])

# ------------------ Handling Outliers ------------------

num_cols = adult.select_dtypes(include=['int64','float64']).columns
z_scores = np.abs(stats.zscore(adult[num_cols]))
adult = adult[(z_scores < 3).all(axis=1)]

# ------------------ Data Transformation ------------------

# Min-Max Scaling
minmax = MinMaxScaler()
adult_minmax = pd.DataFrame(minmax.fit_transform(adult),
                             columns=adult.columns)

# Standard Scaling
standard = StandardScaler()
adult_standard = pd.DataFrame(standard.fit_transform(adult),
                                columns=adult.columns)

print("Adult income preprocessing completed")
```
